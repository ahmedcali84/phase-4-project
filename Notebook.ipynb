## 4. Approaches

### I. Collaborative Filtering

Collaborative filtering is the process of filtering for information or patterns using techniques involving collaboration among multiple agents, viewpoints, data sources. Basically, it is a method of making automatic predictions (filtering) about the interests of a user by collecting preferences or taste information from many users.

There are 2 approaches to Collaborative Filtering:

1) Memory-Based Collaborative Filtering - It is an approach which finds similarity between users or between items to recommend similar items. Examples include Neighbourhood-based CF and Item-based/User-based top-N recommendations.

2) Model-Based Collaborative Filtering - In this approach we use different data mining, machine learning algorithms to predict users' rating of unrated items. Examples include Singular Value Decomposition (SVD).



#### 4.1 Memory-based Collaborative Filtering

```python
# Creating user-item matrix
user_item = df.pivot_table(index = 'userId', columns = 'title', values = 'rating')
```

```python
user_item.head()
```


##### User-based CF


```python
data = user_item.copy()
data = data.fillna(0)
```

```python
data
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity
item_similarity = cosine_similarity(data)

def get_top_recommendations(user_index, data, item_similarity, num_recommendations=5):
    
    user = data.iloc[user_index, :].values
    similar_users = item_similarity[user_index].argsort()[:-num_recommendations-1:-1]

    unrated_movies = np.where(user == 0)[0]

    recommendations = []
    for movie_index in unrated_movies:
        rating = sum(data.iloc[similar_users, movie_index]) / len(similar_users)
        recommendations.append((movie_index, rating))

    
    recommendations.sort(key=lambda x: x[1], reverse=True)

    # Return the top N recommendations
    return recommendations[:num_recommendations] 


user_index_to_recommend = 15  
recommendations = get_top_recommendations(user_index_to_recommend, data, item_similarity)


for movie_index, predicted_rating in recommendations:
    print(f"User {user_index_to_recommend} might like movie {movie_index} with predicted rating {predicted_rating:.2f}")
```

##### item-based 

```python
item = user_item.copy()
```

```python
item = item.fillna(0)
```

```python
def standardize(row):
    new = (row - row.mean()) / (row.max() - row.min())
    return new

data = item.apply(standardize)

# Transpose the matrix to get item-item CF
item_similarity = cosine_similarity(data.T)

new_item = pd.DataFrame(item_similarity,index=item.columns,columns=item.columns)
new_item
```

```python
def similar_movies(movie,rating):
    similar_ratings = new_item[movie]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    return similar_ratings
def recommend():
    return similar_movies("'71 (2014)",5).sort_values(ascending=False)
recommend()
```

#### 4.2 Model-based Collaborative Filtering

```python
from surprise import Reader, Dataset

reader = Reader()
data = Dataset.load_from_df(ratings,reader)
```

```python
dataset = data.build_full_trainset()
print('Number of users: ', dataset.n_users, '\n')
print('Number of items: ', dataset.n_items)
```

#### 4.3 Surprise Library
##### We will be using distance based metrics such as knns and matrix factorization algorithms such as svd which come in surprise library. 

```python
from surprise.prediction_algorithms import SVD
from surprise.prediction_algorithms import KNNWithMeans, KNNBasic, KNNBaseline
from surprise.model_selection import cross_validate
```

##### 1.KNN Basic

```python
# KNNBasic
sim_cos = {"name": "cosine", "user_based": False}

basic = KNNBasic(sim_options=sim_cos)
cv_results = cross_validate(basic, data=data, measures=['RMSE'], cv=5, verbose=True)
```

```python
print('Average RMSE:', np.mean(cv_results['test_rmse']))
```

```python
from surprise.model_selection import GridSearchCV
```

```python
# Define the parameter grid for GridSearchCV
param_grid = {'k': [10, 20,], 
              'sim_options': {'name': ['cosine', 'pearson'],
                              'user_based': [True, False]}}



grid = GridSearchCV(KNNBasic, param_grid, measures=['RMSE', 'MAE'], cv=3)

grid.fit(data)


```

```python
print("Best Parameters: ", grid.best_params['rmse'])
print("Best RMSE: ", grid.best_score['rmse'])
```

##### 2.KNN Baseline 

```python
param_grid = {'k': [20, 30, 40], 
              'sim_options': {'name': ['cosine', 'pearson'],
                              'user_based': [True, False]}}



grid = GridSearchCV(KNNBaseline, param_grid, measures=['RMSE', 'MAE'], cv=3)

grid.fit(data)
```

```python
print("Best Parameters: ", grid.best_params['rmse'])
print("Best RMSE: ", grid.best_score['rmse'])
```

##### 3.KNN WithMeans

```python
param_grid = {'k': [20, 30, 40],
              'sim_options': {'name': ['cosine', 'pearson'],
                              'user_based': [True, False]}}


grid = GridSearchCV(KNNWithMeans, param_grid, measures=['RMSE', 'MAE'], cv=3)

grid.fit(data)
```

```python
print("Best Parameters: ", grid.best_params['rmse'])
print("Best RMSE: ", grid.best_score['rmse'])
```

##### 4.SVD(Singular Value Decomposition)

```python
param_grid = {'n_factors': [5, 10, 15],
              'reg_all': [0.02, 0.1, 0.5]}


grid_search = GridSearchCV(SVD, param_grid, measures=['RMSE', 'MAE'], cv=3)

grid_search.fit(data)
```

```python
print("Best Parameters: ", grid_search.best_params['rmse'])
print("Best RMSE: ", grid_search.best_score['rmse'])
```
